```python
#  Visualize the locations of restaurants on a map using latitude and longitude information.

import folium
import pandas as pd
import folium
# Initialize a map centered around a specific point
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=5)

# Add markers for each restaurant
for i, j in zip(df['Latitude'], df['Longitude']):
    folium.Marker(location=[i, j]).add_to(m)

# Display the map
m
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway




# Basic Correlation Analysis
correlation_matrix = df[['Latitude', 'Longitude', 'Aggregate rating']].corr()
print(correlation_matrix)

# Visualization
sns.scatterplot(x='Latitude', y='Aggregate rating', data=df)
plt.title('Latitude vs. Rating')
plt.show()

sns.scatterplot(x='Longitude', y='Aggregate rating', data=df)
plt.title('Longitude vs. Rating')
plt.show()

sns.boxplot(x='City', y='Aggregate rating', data=df)
plt.title('City vs. Rating')
plt.show()

# Advanced Analysis (ANOVA)
city_groups = [group['Aggregate rating'].values for Locality, group in df.groupby('City')]
anova_result = f_oneway(*city_groups)
print('ANOVA result:', anova_result)
```

```python
#Extract additional features from the existing columns, such as the length of the restaurant name or address.

# Extract the length of the restaurant name
df['Restaurant Name Length'] = df['Restaurant Name'].apply(len)

# Extract the length of the address
df['Address Length'] = df['Address'].apply(len)


```

```python
#  Create new features like "Has Table Booking"or "Has Online Delivery" by encoding categorical variables.




# Convert 'yes'/'no' to 1/0 for 'Has Table booking'
df['Has Table booking'] = df['Has Table booking'].map({'yes': 1, 'no': 0})

# Convert 'yes'/'no' to 1/0 for 'Has Online delivery'
df['Has Online delivery'] = df['Has Online delivery'].map({'yes': 1, 'no': 0})

df.head()
```
